```python
import requests
```

```python
try:
    response = requests.get("http://www.law.go.kr/DRF/lawSearch.do", params={
        "OC": "yoonjs1459",
        "target": "law",
        "type": "JSON",
        "search": 2,
        "display": 5,
        "query": "민법"
    })
    response.raise_for_status()
    data = response.json()
    print(data)
except requests.exceptions.RequestException as e:
    print(f"API 요청 중 오류 발생: {e}")
```

```python
print(response.text)
```
